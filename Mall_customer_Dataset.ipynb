## Determine Optimal Number of Clusters (K)

### Subtask:
Use the Elbow Method to determine the optimal number of clusters (K) for the K-means algorithm.


**Reasoning**:
I need to implement the Elbow Method to determine the optimal number of clusters. This involves importing necessary libraries, iterating through a range of cluster numbers to calculate WCSS, and then plotting these values to identify the 'elbow point'.



````python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

wcss = []

# Loop through a range of possible cluster numbers (e.g., from 1 to 10 or 15)
for i in range(1, 11): # Commonly, K is tested up to 10 or 15
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init='auto')
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Plot the WCSS values against the range of cluster numbers
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.grid(True)
plt.xticks(range(1, 11))
plt.show()

print("Elbow plot generated to help determine the optimal number of clusters (K).")
````

````python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
from google.colab import files

# Check if the file exists, if not, prompt for upload
try:
    df = pd.read_csv('/content/Mall_Customers.csv')
    print("Mall_Customers.csv found locally.")
except FileNotFoundError:
    print("Mall_Customers.csv not found. Please upload the file.")
    uploaded = files.upload()
    for fn in uploaded.keys():
        print(f'User uploaded file "{fn}"')
    # After upload, try loading again. Assuming the uploaded file is named 'Mall_Customers.csv'
    df = pd.read_csv('/content/Mall_Customers.csv')

# Select the specified features
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

wcss = []

# Loop through a range of possible cluster numbers (from 1 to 10)
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init='auto')
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Plot the WCSS values against the range of cluster numbers
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.grid(True)
plt.xticks(range(1, 11))
plt.show()

print("Elbow plot generated to help determine the optimal number of clusters (K).")
````
